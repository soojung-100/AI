KNN (MNIST활용)


```
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data", one_hot=True)
import tensorflow as tf
import numpy as np

n_train_img = 5000
n_test_img = 200
pixel_train, onehot_train = mnist.train.next_batch(n_train_img)
pixel_test, onehot_test = mnist.train.next_batch(n_test_img)

TRAIN = tf.placeholder("float", [None, 784] )
TEST = tf.placeholder("float", [784] )  

#맨하탄 디스턴스
d = tf.reduce_sum(tf.abs(tf.add(TRAIN, tf.negative(TEST))), reduction_indices=1)   #tb.abs (절대값)    |x1 - x2| + |y1 - y2| + |z1 - z2|......, reduction_indices는 axis랑 동일의미
#Euclidean 디스턴스
#d =  tf.sqrt(tf.reduce_sum(tf.square(tf.abs(tf.add(TRAIN, tf.negative(TEST)))), reduction_indices=1))

val, idx = tf.nn.top_k(-d, k=5, sorted=False) 
acc = 0.0

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(n_test_img):
        knn_idx = sess.run(idx, feed_dict={TRAIN: pixel_train, TEST: pixel_test[i,:]})  #거리가 가까운 5개의 index [3580 1097 4597 2554 3177] ....test수만큼
        look_up = np.zeros(10)
        for j in np.argmax(onehot_train[knn_idx], axis=1):  # np.argmax 최대값 색인찾기
            look_up[j] += 1
        pred = np.argmax(look_up)

        print("Test:", i)
        print("Pred:", pred, "Real:" , np.argmax(onehot_test[i]))
        if pred == np.argmax(onehot_test[i]):
            acc += 1./n_test_img           #####?????          
    print("Accuracy:", acc)
```
